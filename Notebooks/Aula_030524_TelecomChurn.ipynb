# Pre-process

```python
columnToDrop = [
    'phone number',
    'total day charge',
    'total eve charge',
    'total night charge',
    'total intl charge']
df=df.drop(columnToDrop,axis=1)
df
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
enc=LabelEncoder()

colunaParaCodificar=['state','international plan','voice mail plan','churn']
for c in colunaParaCodificar:
    df[c]=enc.fit_transform(df[c])
    print(enc.classes_)
```

# Split

```python
y=df['churn']
X=df.drop(['churn'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=24,stratify=y)
```
