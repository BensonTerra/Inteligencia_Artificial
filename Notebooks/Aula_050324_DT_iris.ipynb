```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
```

```python
dados = load_iris()
dados
```

```python
print(dados.DESCR)
```

```python
X = dados.data
y = dados.target
```

# Preparar Dados

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24, test_size=0.25, stratify=y)
```

# criar modelo de calssificação

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

# Fazer cross-validation

```python
scores = cross_val_score(clf, X, y, cv=5)
scores
```

# Treinar o classificador

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

#imprimir a precisão da previsão com dados de treino e de teste
print("Acurácia com dados de treino:",accuracy_score(y_true=y_train, y_pred=clf.predict(X_train)))
print("Acurácia com dados de treino:",accuracy_score(y_true=y_test, y_pred=clf.predict(X_test)))
```
